```python
import matplotlib.pyplot as plt
import numpy as np

def get_gauss(mu, sigma, peak_height=None):
    if peak_height is None:
        scale = (2*np.pi*sigma**2)**-0.5
    else:
        scale = peak_height
    def gauss(x):
        return scale*np.exp(-(x-mu)**2/sigma**2)
    
    return gauss

p1 = get_gauss(1,0.1, 0.2)
p2 = get_gauss(1,1,0.2)
f = lambda x: x**0.5
x = np.linspace(0,3, 2000)
```

```python
mosaic = [
    ["py1", "f1",  ".", "py2", "f2"],
    [".", "px1", ".",  ".",  "px2"],
]

%config InlineBackend.figure_formats = ['svg']
fig, axd = plt.subplot_mosaic(mosaic=mosaic, 
                              width_ratios=[0.2, 1, 0.5, 0.2, 1],
    height_ratios=[1., 0.2,], empty_sentinel=".")

ax_px1 = axd["px1"]
ax_px2 = axd["px2"]
ax_f1 = axd["f1"]
ax_py1 = axd["py1"]
ax_py2 = axd["py2"]
ax_f2 = axd["f2"]

ax_f1.set_aspect("auto")
ax_f2.set_aspect("auto")

cmap = plt.cm.get_cmap("Greys")
from matplotlib.colors import LinearSegmentedColormap
cmap = LinearSegmentedColormap.from_list("lim_Greys", cmap(np.linspace(0.2, 1, 1000)), 1000)

def plot_with_color(ax, x, y, c):
    points = np.array([x, y]).T.reshape(-1, 1, 2)
    segments = np.concatenate([points[:-1], points[1:]], axis=1)
    norm = plt.Normalize(c.min(), c.max())
    from matplotlib.collections import LineCollection
    
    lc = LineCollection(segments, cmap=cmap, norm=norm)
    lc.set_array(c)
    lc.set_linewidth(2)

    line = ax.add_collection(lc)
    ax.autoscale_view()

cx = "tab:blue"
cy = "tab:green"
ax_px1.plot(x, p1(x), c=cx)
plot_with_color(ax_f1, x, f(x), p1(x))


ax_px2.plot(x, p2(x), c=cx)
plot_with_color(ax_f2, x, f(x), p2(x))

dx = np.diff(x)[0]
def p_y(y, f, p):
    py = np.mean(p(x)[np.abs(f(x)-y)<=f(dx)])
    return py

p_y = np.vectorize(p_y, excluded=["f", "p"])

y = f(x)

py = p_y(y, f, p1)
ax_py1.plot(py/max(py)*1, y, c=cy)

py = p_y(y, f, p2)
ax_py2.plot(py/max(py)*1, y, c=cy)

for ax in axd.values():
    ax.set_xticks([])
    ax.set_yticks([])
    ax.margins(0.2, 0.2)
    
ax_py1.set_ylabel("$y=f(x)$")
ax_py2.set_ylabel("$y=f(x)$")
ax_px1.set_xlabel("$x$")
ax_px2.set_xlabel("$x$")
```
